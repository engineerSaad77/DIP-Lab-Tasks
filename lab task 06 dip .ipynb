```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# Load the image in grayscale
image = cv2.imread('/Dip_Tasks/cat.jpg', cv2.IMREAD_GRAYSCALE)

# Display the original image
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Compute the 2D FFT
fft_image = np.fft.fft2(image)

# Shift the zero frequency component to the center
fft_shifted = np.fft.fftshift(fft_image)
```

```python
# Compute magnitude spectrum
magnitude_spectrum = np.log(np.abs(fft_shifted) + 1)  # Adding 1 to avoid log(0)

# Compute phase spectrum
phase_spectrum = np.angle(fft_shifted)

# Display Magnitude Spectrum
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum')
plt.axis('off')

# Display Phase Spectrum
plt.subplot(1, 2, 2)
plt.imshow(phase_spectrum, cmap='gray')
plt.title('Phase Spectrum')
plt.axis('off')
plt.show()
```

```python
# Create a low-pass filter mask
rows, cols = image.shape
crow, ccol = rows // 2, cols // 2
mask = np.zeros((rows, cols), np.uint8)

# Set the radius for the low pass filter
r = 30
cv2.circle(mask, (ccol, crow), r, 1, thickness=-1)

# Apply mask to FFT image
low_freq_fft = fft_shifted * mask

# Create a high-pass filter mask
high_mask = 1 - mask
high_freq_fft = fft_shifted * high_mask
```

```python
# Inverse FFT for low frequencies
low_freq_image = np.fft.ifft2(np.fft.ifftshift(low_freq_fft))
low_freq_image = np.abs(low_freq_image)  # Take the magnitude

# Inverse FFT for high frequencies
high_freq_image = np.fft.ifft2(np.fft.ifftshift(high_freq_fft))
high_freq_image = np.abs(high_freq_image)
```

```python
# Display the results
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(low_freq_image, cmap='gray')
plt.title('Low-Frequency Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(high_freq_image, cmap='gray')
plt.title('High-Frequency Image')
plt.axis('off')
plt.show()
```
